```python
Barrios = df_filtro.groupby('l3').mean().sort_values('price', ascending=False).index
plt.figure(figsize=(12,10))
sns.barplot(x='l3', y='price', data=df_filtro, order=Barrios, linewidth=1.1, edgecolor='black')
plt.xticks(rotation=90)
plt.title('Precio por barrios Capital Fedral', size=25)
plt.xlabel('Barrio', size= 15)
plt.ylabel('Precio en US$', size= 15)
plt.grid()
plt.show()
```
